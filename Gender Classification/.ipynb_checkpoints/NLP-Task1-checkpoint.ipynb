```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() 
```

```python
# file path - this for linux windows you will need "//"
f_path = "cardio_train.csv"
'''
record ID - The unique identifier for each connection record.
duration_  This feature denotes the number of seconds (rounded) of the connection. For example, a connection for 0.17s or 0.3s would be indicated with a “0” in this field.
src_bytes This field represents the number of data bytes transferred from the source to the destination (i.e., the amount of out-going bytes from the host).
dst_bytes This feature represents the number of data bytes transferred from the destination to the source (i.e., the amount of bytes received by the host).
'''
df = pd.read_csv(f_path,sep = ';',index_col=0)
```

```python
df2 = df[['age','gender','height','weight','smoke']]
df2 = df2.head(50000)
df2['gender'] = df2['gender'].replace({1:0,2:1})
```

```python
df2,df2['gender'].value_counts()
```

```python
features = 4
alpha = 0.00001
# placeholders are a bunch of empty nodes in a graph, waiting to be filled
# None -> allow num of examples to be dynamic size
# X and Y represnt matrices
X = tf.placeholder(tf.float32,[None,features])
Y = tf.placeholder(tf.float32, [None, 1])

# Trainable Variable Weights
W = tf.Variable(tf.zeros([features,1]))
  
# Trainable Variable Bias
b = tf.Variable(tf.zeros([1]))
```

```python
# Hypothesis
y_ = tf.nn.sigmoid(tf.add(tf.matmul(X, W), b))
  
# Sigmoid Cross Entropy Cost Function
loss = tf.nn.sigmoid_cross_entropy_with_logits(
                    logits = y_, labels = Y)
  
# Gradient Descent Optimizer
update = tf.train.GradientDescentOptimizer(alpha).minimize(loss)


data_x = df2.drop(columns='gender').to_numpy()
data_y = df2.drop(columns=['age', 'weight','height','smoke']).to_numpy()
```

```python
# confirm dimensions
data_x, data_y
```

```python
physical_devices = tf.config.list_physical_devices('GPU') 
for device in physical_devices:
    tf.config.experimental.set_memory_growth(device, True)
```

# OPTION 1 - based on Amos

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for i in range(0,500000):
    sess.run(update,feed_dict={X: data_x, Y : data_y})
    if i%10000 == 0:
        print('Iteration: ',i)
        print('W:\n ',sess.run(W), '\nb:\n ', sess.run(b), '\nLoss:\n ',loss.eval(session=sess, feed_dict={X:data_x,Y:data_y})[0])
print('FW:\n ',sess.run(W), '\nFb:\n ', sess.run(b), '\nFLoss:\n ',loss.eval(session=sess, feed_dict={X:data_x,Y:data_y})[0])
```

```python
# Imports
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() 
```

```python
# file path - this for linux windows you will need "//"
f_path = "cardio_train.csv"
'''
record ID - The unique identifier for each connection record.
duration_  This feature denotes the number of seconds (rounded) of the connection. For example, a connection for 0.17s or 0.3s would be indicated with a “0” in this field.
src_bytes This field represents the number of data bytes transferred from the source to the destination (i.e., the amount of out-going bytes from the host).
dst_bytes This feature represents the number of data bytes transferred from the destination to the source (i.e., the amount of bytes received by the host).
'''
df = pd.read_csv(f_path,sep = ';',index_col=0)
```

```python
df2 = df[['age','gender','height','weight','smoke']]
df2 = df2.head(50000)
df2['gender'] = df2['gender'].replace({1:0,2:1})
```

```python
df2,df2['gender'].value_counts()
```

```python
features = 4 
hidden_layer_node = 10

X = tf.placeholder(tf.float32,[None,features])
Y = tf.placeholder(tf.float32,[None,1])

#matrix, dimensions of W  
W1 = tf.Variable(tf.truncated_normal([features,hidden_layer_node],stddev = 0.1))
#vector
b1 = tf.Variable(tf.constant(0.1,shape=[hidden_layer_node]))
#
Z1 = tf.add(tf.matmul(X,W1),b1)
#ReLu activation layer
a1 = tf.nn.relu(Z1)
# 50 examples with 2 features --> 50 examples with 10 features
W2 = tf.Variable(tf.truncated_normal([hidden_layer_node,1],stddev = 0.1))
#
b2 = tf.Variable(0.)
#
Z2 = tf.matmul(a1,W2) + b2
#
sigmoid = 1 / (1.0 + tf.exp(-Z2))
#
loss = tf.reduce_mean(-(Y* tf.log(sigmoid) + (1-Y)*tf.log(1-Y)))
#
update = tf.train.GradientDescentOptimizer(0.0001).minimize(loss)

data_x = df2.drop(columns='gender').to_numpy()
data_y = df2.drop(columns=['age', 'weight','height','smoke']).to_numpy(dtype=np.float32)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for i in range(0,50000):
    sess.run(update,feed_dict={X:data_x, Y:data_y})
```

```python
print("Prediction:",sigmoid.eval(session=sess,feed_dict={X:[[20228,156,85.0,0],[18393,168,62.0,0],[18857,165,64.0,0],[ 17623,169,82.0,0]]}))
```
